```python
# Dependencies
import os
import pandas as pd
import numpy
```

```python
# File to load
data_file = os.path.join("Resources","Pokemon.csv")
```

```python
# Read with Pandas
data_frame = pd.read_csv(data_file)
```

```python
# Extract the following columns: "Type 1", "HP", "Attack", "Sp. Atk", "Sp. Def", and "Speed"
#data_frame.count()
new_data_frame = data_frame[["Type 1","HP","Attack","Sp. Atk","Sp. Def","Speed"]]
new_data_frame.head()
```

```python
# Create a dataframe of the average stats for each type of pokemon.
avg_data_frame = new_data_frame.groupby("Type 1").sum() / new_data_frame.groupby("Type 1").count()
avg_data_frame
```

```python
# Calculate the total power level of each type of pokemon by adding all of the stats together.
# Place the results into a new column.
avg_data_frame['Summed Power'] = avg_data_frame[['HP','Attack','Sp. Atk','Sp. Def','Speed']].sum(axis = 1)
avg_data_frame
```

```python
# Bonus: Sort the table by strongest type and export the resulting table to a new CSV.

sorted_frame = avg_data_frame.sort_values(["Summed Power"])
sorted_frame
```
